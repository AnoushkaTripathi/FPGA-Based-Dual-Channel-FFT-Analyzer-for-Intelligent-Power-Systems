```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# Step 1: Generate synthetic satellite power consumption data
def generate_synthetic_power_data(length=1024, noise_level=0.05):
    t = np.linspace(0, 1, length)
    base_signal = np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 120 * t)
    noise = np.random.normal(0, noise_level, length)
    return base_signal + noise

# Step 2: FFT of the signal
signal = generate_synthetic_power_data()
frequency_spectrum = fft(signal)

# Step 3: Add random harmonics (simulate anomalies)
def add_random_harmonics(frequency_spectrum, num_harmonics, magnitude):
    spectrum_with_harmonics = frequency_spectrum.copy()
    n = len(spectrum_with_harmonics)
    for _ in range(num_harmonics):
        harmonic_index = np.random.randint(1, n//2)
        spectrum_with_harmonics[harmonic_index] += magnitude
    return spectrum_with_harmonics

frequency_spectrum = add_random_harmonics(frequency_spectrum, num_harmonics=5, magnitude=100)

# Step 4: Inverse FFT
def inverse_fft_transform(frequency_spectrum):
    return np.real(ifft(frequency_spectrum))

# Step 5: Autoencoder for anomaly detection
def build_autoencoder(input_dim, encoding_dim=64, learning_rate=0.001):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(64, activation='relu')(input_layer)
    decoded = Dense(input_dim, activation='linear')(encoded)
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer=Adam(learning_rate=learning_rate), loss='mean_squared_error')
    return autoencoder

# Prepare training data
frequency_spectrum_real = np.real(frequency_spectrum)
input_dim = len(frequency_spectrum_real)
frequency_spectrum_real_reshaped = frequency_spectrum_real.reshape(1, -1)

# Train autoencoder
autoencoder = build_autoencoder(input_dim, encoding_dim=64, learning_rate=0.0001)
autoencoder.fit(frequency_spectrum_real_reshaped, frequency_spectrum_real_reshaped,
                epochs=813, batch_size=1, shuffle=True)

# Reconstruct signal
reconstructed_spectrum = autoencoder.predict(frequency_spectrum_real_reshaped)

# Calculate error
mae = np.mean(np.abs(frequency_spectrum_real - reconstructed_spectrum.flatten()))
print("Mean Absolute Error:", mae)

# Anomaly detection
percentage_error = np.abs((frequency_spectrum_real - reconstructed_spectrum.flatten()) / frequency_spectrum_real) * 100
threshold = 10  # adjustable
valid_indices = np.where(percentage_error > threshold)[0]
invalid_indices = np.where(percentage_error <= threshold)[0]

# Visualization
plt.figure(figsize=(12, 6))
plt.plot(frequency_spectrum_real, label='Original Spectrum')
plt.plot(reconstructed_spectrum.flatten(), label='Reconstructed Spectrum')
plt.scatter(valid_indices, frequency_spectrum_real[valid_indices], color='red', label='Anomalies')
plt.title('Power Spectrum with Anomaly Detection')
plt.xlabel('Frequency Bins')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Detected Anomalies:", len(valid_indices))
```
